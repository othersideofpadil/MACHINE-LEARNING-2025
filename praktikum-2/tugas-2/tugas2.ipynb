```python
import pandas as pd
```

```python
df = pd.read_csv("../data/day.csv")
```

```python
df.head()
```

```python
total_data = len(df)
```

```python
train_size = int(0.8 * total_data) #80% data training
test_size = total_data - train_size #20% data testing
val_size = int(0.1 * train_size) #10% dari training untuk validasi
```

```python
train_full = df[:train_size]         # 80% pertama untuk training
test_df = df[train_size:]            # 20% terakhir untuk testing
val_df = train_full[:val_size]       # 10% awal dari training jadi validation
train_df = train_full[val_size:]     # sisa training
```

```python
# menampilkan jumlah data
print("Jumlah total data:", total_data)
print("Jumlah data Training:", len(train_df))
print("Jumlah data Validation:", len(val_df))
print("Jumlah data Testing:", len(test_df))

print("\n--- Data Training (5 teratas) ---")
print(train_df.head())

print("\n--- Data Validation (5 teratas) ---")
print(val_df.head())

print("\n--- Data Testing (5 teratas) ---")
print(test_df.head())
```
